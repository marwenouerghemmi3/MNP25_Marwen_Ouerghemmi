```python
import numpy as np
import matplotlib.pyplot as plt
import os

#  Paramètres physiques
kB = 1.38e-23        # Constante de Boltzmann (J/K)
T = 300              # Température (K)
kappa = 0.2e-6       # Raideur de la pince optique (N/m)
gamma = 1e-8         # Coefficient de friction (kg/s)
dt = 1e-4            # Pas de temps (s)
n_steps = 300000     # Nombre d'itérations
sigma = 2e-6         # Paramètre Lennard-Jones (m)
epsilon = 6 * kB * T # Profondeur LJ (J)
# Plusieurs distances initiales r_0
r_0_values = [1.8e-6, 2.0e-6, 2.2e-6, 2.5e-6, 2.8e-6, 3.0e-6, 3.2e-6]
bins = 150           # Nombre de bins pour histogrammes

# Seuils pour filtrage des bins
min_counts = 10      # Nombre minimum d'occurrences par bin
rel_threshold = 0.02 # Densité relative minimale
smooth_sigma = 1.0   # Lissage des histogrammes (bins)

# Dossier pour sauvegarder figures et fichiers CSV
fig_dir = "figures"
os.makedirs(fig_dir, exist_ok=True)

#  Fonctions physiques
def F_trap(x, x0):
    """Force harmonique de la pince optique"""
    return -kappa * (x - x0)

def F_LJ(r):
    """Force Lennard-Jones"""
    return 24 * epsilon * (2*(sigma/r)**12 - (sigma/r)**6) / r

def U_LJ_theoretical(r):
    """Potentiel Lennard-Jones"""
    return 4 * epsilon * ((sigma / r)**12 - (sigma / r)**6)

# Simulation Langevin
def simulate_langevin(r_0, with_pair=True, verbose=False):
    x0_1, x0_2 = 0.0, r_0
    x = np.array([x0_1, x0_2])
    distances = np.empty(n_steps)

    for step in range(n_steps):
        Ftrap = np.array([F_trap(x[0], x0_1), F_trap(x[1], x0_2)])
        if with_pair:
            r = x[1] - x[0]
            Fpair = np.array([-F_LJ(r), F_LJ(r)])
        else:
            Fpair = np.array([0.0, 0.0])
        noise = np.sqrt(2*kB*T*gamma/dt) * np.random.randn(2)
        x += dt/gamma * (Ftrap + Fpair) + noise*dt/gamma
        distances[step] = x[1] - x[0]
        if verbose and (step % 50000 == 0):
            print(f"Step {step}/{n_steps}, r={x[1]-x[0]:.3e} m")
    return distances

#  Lissage
def gaussian_kernel(size, sigma):
    L = np.arange(-size, size+1)
    kernel = np.exp(-(L**2)/(2*sigma**2))
    kernel /= np.sum(kernel)
    return kernel

def smooth_hist(hist, sigma):
    if sigma <= 0:
        return hist
    size = int(max(1, np.ceil(3*sigma)))
    kernel = gaussian_kernel(size, sigma)
    hist_padded = np.pad(hist, pad_width=size, mode='reflect')
    smoothed = np.convolve(hist_padded, kernel, mode='same')
    return smoothed[size:-size]

#  Analyse pour chaque r_0
Upair_all = []

for i, r_0 in enumerate(r_0_values):
    print(f"\n--- Traitement pour r_0 = {r_0*1e6:.2f} µm ---")

    # Simulation sans interaction
    dist_no_pair = simulate_langevin(r_0, with_pair=False, verbose=True)

    # Simulation avec interaction LJ
    dist_with_pair = simulate_langevin(r_0, with_pair=True, verbose=True)

    # Histogrammes
    r_min = min(np.min(dist_no_pair), np.min(dist_with_pair))
    r_max = max(np.max(dist_no_pair), np.max(dist_with_pair))
    edges = np.linspace(r_min, r_max, bins+1)

    counts0, _ = np.histogram(dist_no_pair, bins=edges)
    countsU, _ = np.histogram(dist_with_pair, bins=edges)
    width = edges[1] - edges[0]

    rho0 = counts0 / (np.sum(counts0)*width)
    rhoU = countsU / (np.sum(countsU)*width)
    r_vals = 0.5*(edges[:-1]+edges[1:])

    # Lissage
    rho0_s = smooth_hist(rho0, smooth_sigma)
    rhoU_s = smooth_hist(rhoU, smooth_sigma)

    # Domaine fiable pour éviter ln(0)
    max_rho0 = np.max(rho0_s)
    max_rhoU = np.max(rhoU_s)
    mask_counts = (counts0 >= min_counts) & (countsU >= min_counts)
    mask_rel = (rho0_s >= rel_threshold*max_rho0) & (rhoU_s >= rel_threshold*max_rhoU)
    mask = mask_counts & mask_rel

    if np.any(mask):
        indices = np.where(mask)[0]
        groups = np.split(indices, np.where(np.diff(indices)!=1)[0]+1)
        groups_sorted = sorted(groups, key=lambda g: len(g), reverse=True)
        best_group = groups_sorted[0]
        mask_final = np.zeros_like(mask, dtype=bool)
        mask_final[best_group] = True
        mask = mask_final
    else:
        mask = (rho0_s > 0) & (rhoU_s > 0)

    r_vals_valid = r_vals[mask]
    rho0_valid = rho0_s[mask]
    rhoU_valid = rhoU_s[mask]

    # Reconstruction du potentiel
    eps_small = 1e-30
    Upair_rec = -kB*T*(np.log(rhoU_valid+eps_small) - np.log(rho0_valid+eps_small))

    # Ajustement pour comparaison LJ
    Upair_scaled = Upair_rec / epsilon
    U_theo_scaled = U_LJ_theoretical(r_vals_valid) / epsilon
    min_index = np.argmin(U_theo_scaled)
    Upair_final = Upair_scaled - Upair_scaled[min_index] + U_theo_scaled[min_index]

    Upair_all.append((r_vals_valid, Upair_final))

    # Histogrammes
    plt.figure(figsize=(6,4))
    plt.plot(r_vals*1e6, rho0_s, 'b--', label='rho0 (sans LJ)')
    plt.plot(r_vals*1e6, rhoU_s, 'r-', label='rhoU (avec LJ)')
    plt.fill_between(r_vals*1e6, 0, np.maximum(rho0_s, rhoU_s), where=mask, color='yellow', alpha=0.3, label='Zone fiable')
    plt.xlabel("Distance r [µm]")
    plt.ylabel("Densité normalisée")
    plt.title(f"Histogrammes pour r_0={r_0*1e6:.2f} µm")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"{fig_dir}/histograms_r0_{i}.png", dpi=200)
    plt.show()

# Comparaison 2 par 2
for i, (r_vals_valid, Upair_final) in enumerate(Upair_all):
    plt.figure(figsize=(6,4))
    plt.plot(r_vals_valid*1e6, Upair_final, 'r-', label='Reconstruit')
    plt.plot(r_vals_valid*1e6, U_LJ_theoretical(r_vals_valid)/epsilon, 'b--', label='Théorique LJ')
    plt.xlabel("Distance r [µm]")
    plt.ylabel(r"$U_{pair}/\epsilon$")
    plt.title(f"Potentiel pour r_0={r_0_values[i]*1e6:.2f} µm")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig(f"{fig_dir}/Upair_vs_LJ_r0_{i}.png", dpi=200)
    plt.show()

# Graphique global pour tous les r_0
plt.figure(figsize=(7,5))
colors = plt.cm.viridis(np.linspace(0,1,len(r_0_values)))
for i, (r_vals_valid, Upair_final) in enumerate(Upair_all):
    plt.plot(r_vals_valid*1e6, Upair_final, color=colors[i], label=f"Reconstruit r_0={r_0_values[i]*1e6:.1f} µm")
    plt.plot(r_vals_valid*1e6, U_LJ_theoretical(r_vals_valid)/epsilon, '--', color=colors[i], label=f"LJ r_0={r_0_values[i]*1e6:.1f} µm")
plt.xlabel("Distance r [µm]")
plt.ylabel(r"$U_{pair}/\epsilon$")
plt.title("Comparaison globale : tous les r_0")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig(f"{fig_dir}/Upair_vs_LJ_all_r0.png", dpi=200)
plt.show()
```
